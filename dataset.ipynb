```python
import os
import json
from pprint import pprint
import torch
import torchvision
import numpy as np
import pandas as pd
from PIL import Image
from torch.utils.data import Dataset
from torchvision.transforms.functional import pil_to_tensor
```

```python
with open(os.path.join('hico_20160224_det', 'anno_cleaned.json'), 'r') as fp:
    annotation = json.load(fp=fp)
```

```python
bbox_anno = annotation['bbox_train']
for anno in bbox_anno:
    has_list = False
    for hoi in anno['hoi']:
        for bbox in hoi['bboxhuman']:
            if isinstance(bbox, list):
                has_list = True
        for bbox in hoi['bboxobject']:
            if isinstance(bbox, list):
                has_list = True
    if has_list:
        pprint(anno)
```

```python
bbox_anno = annotation['bbox_train']
bbox_anno = [anno for anno in bbox_anno if not all(hoi['invis'] == 1 for hoi in anno['hoi'])]
len(bbox_anno)
i = 0
for anno in bbox_anno:
    p = False
    for hoi in anno['hoi']:
        if len(hoi['bboxhuman']) == 0:
            p = True
        if len(hoi['bboxobject']) == 0:
            p = True
    if p:
        i += 1
        pprint(anno)
i
```
